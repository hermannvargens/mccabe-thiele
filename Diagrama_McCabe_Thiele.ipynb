Problema:  

    Encontrar o NГєmero de Pratos TeГіricos para uma destilaГ§ГЈo binГЎria, usando o MГ©todo de Mccacbe-Thiele

Dados do Problema

    AlimentaГ§ГЈo (zFвЂ‹): 0,5 (50% do componente mais volГЎtil).

    Destilado desejado (xDвЂ‹): 0,9.

    ResГ­duo desejado (xBвЂ‹): 0,1.

    RazГЈo de Refluxo (R): 2,0.

    CondiГ§ГЈo da Carga (q): LГ­quido saturado (q=1).

## 0 Definir as variГЎveis do problema

```python
xD, zF, xB = 0.90, 0.5, 0.1
R = 2
q = 1
alfa = 2.5
```

## 1 Definir a Curva de EquilГ­brio

A curva de equilГ­brio para sistemas ideais segue a equaГ§ГЈo:

$$y = \frac{\alpha \cdot x}{1 + (\alpha - 1)x}$$

Para $\alpha = 2,5$, se $x = 0,4$, entГЈo:

$$y = \frac{2,5 \cdot 0,4}{1 + (1,5 \cdot 0,4)} = 0,625$$

VocГЄ plota esses pontos para formar a curva acima da diagonal de 45В°.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vamos criar um array com todos os valores de x para os quais queremos encontrar y, e construir a curva de equilГ­brio:

x_eq = np.linspace(0, 1, 100)

# Em seguida, calculamos os valores de y correspondentes:

y_eq = (alfa * x_eq) / (1 + (alfa - 1) * x_eq)

# Vamos ver alguns valores:

print("Valores de x:\n", x_eq[:10])
print("\nValores de y:\n", y_eq[:10])
```

```python
# Agora vamos ver como ficou o grГЎfico da curva de equilГ­brio:

# A curva de equilГ­brio
plt.plot(x_eq, y_eq, label='Curva de EquilГ­brio')
plt.title('Diagrame de McCabe-Thiele')
plt.legend()
plt.grid(True)

#Para visualizarmos tambГ©m a reta diagonal, precisamos criГЎ-la:

plt.plot([0, 1], # (x1вЂ‹,x2вЂ‹)
         [0, 1], # (y1,y2вЂ‹)
         'k--', alpha=0.5)

plt.show()

#TODO: altere o cГіdigo para incluir uma legenda para a linha diagonal
```

A partir daqui, vamos utilizar funГ§Гµes para modularizar cada etapa que estamos adicionando ao grГЎfico.

```python
# FunГ§ГЈo para a Base do GrГЎfico
def plotar_base_equilibrio(x_eq, y_eq):
    plt.plot(x_eq, y_eq, label='Curva de EquilГ­brio')
    plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)
    plt.title('Diagrama de McCabe-Thiele')
    plt.legend()
    plt.grid(True)
    plt.xlim(0, 1)
    plt.ylim(0, 1)
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plt.show()
```

## 2 LocalizaГ§ГЈo das ComposiГ§Гµes

Marque no eixo x (fase lГ­quida) as trГЄs fraГ§Гµes molares principais:

    xDвЂ‹: FraГ§ГЈo desejada no destilado (topo).

    xFвЂ‹: FraГ§ГЈo na alimentaГ§ГЈo (feed).

    xBвЂ‹: FraГ§ГЈo no resГ­duo (fundo).

```python
def plotar_composicoes(xD, zF, xB):

    # Ponto do Destilado (xD) no eixo x
    plt.plot([xD, xD], [0, xD], color='red', linestyle=':', linewidth=1) # linha vertical
    plt.plot(xD, 0, 'ro', clip_on=False) # 'ro' = cГ­rculo vermelho; clip_on=False permite desenhar na borda
    plt.annotate('$x_D$', # para escrever em LaTeX
                 xy=(xD, 0), # coordenada do ponto que estamos apontando
                 xytext=(xD, -0.05), # posiГ§ГЈo do texto
                 ha='center', # horizontal alignment. Alinha o centro do texto com a coordenada fornecida
                 fontweight='bold',# negrito
                 color='red') # cor vermelha

    # Ponto da AlimentaГ§ГЈo (zF) no eixo x
    plt.plot([zF, zF], [0, zF], color='green', linestyle=':', linewidth=1)
    plt.plot(zF, 0, 'go', clip_on=False) # 'go' = cГ­rculo verde
    plt.annotate('$z_F$', xy=(zF, 0), xytext=(zF, -0.05),
                ha='center', fontweight='bold', color='green')

    # Ponto do Fundo (xB) no eixo x
    plt.plot([xB, xB], [0, xB], color='blue', linestyle=':', linewidth=1)
    plt.plot(xB, 0, 'bo', clip_on=False) # 'bo' = cГ­rculo azul
    plt.annotate('$x_B$', xy=(xB, 0), xytext=(xB, -0.05),
                ha='center', fontweight='bold', color='blue')

    # --- LEGENDAS DOS EIXOS ---

    plt.xlabel('FraГ§ГЈo molar no lГ­quido (x)')
    plt.ylabel('FraГ§ГЈo molar no vapor (y)')
    plt.grid(True)
    plt.legend()
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plotar_composicoes(xD, zF, xB)
plt.show()
```

## 3 Linha de OperaГ§ГЈo da SeГ§ГЈo de RetificaГ§ГЈo (LOR)

Esta linha representa o que acontece acima do prato de alimentaГ§ГЈo. Ela parte do ponto $(x_D, x_D)$ na linha de 45В°.

A equaГ§ГЈo Г©:

$$y = \frac{R}{R+1}x + \frac{x_D}{R+1}$$

Onde $R$ Г© a razГЈo de refluxo. VocГЄ pode traГ§ГЎ-la ligando $(x_D, x_D)$ ao intercepto no eixo $y$ dado por $\frac{x_D}{R+1}$.

```python
def plotar_LOR(R):

  # 1. Calculando o intercepto no eixo y (ponto onde x = 0)
  intercepto_y = xD / (R + 1)

  # 2. Plotando a LOR
  # Vamos ligar o ponto (0, intercepto_y) atГ© o ponto (xD, xD)
  plt.plot([0, xD], [intercepto_y, xD], color='red', linewidth=1, label='LOR (RetificaГ§ГЈo)')
  plt.legend()

  # 3. Marcando o ponto do intercepto
  plt.plot(0, intercepto_y, 'ko') # Ponto preto no eixo y
  plt.annotate(f'Intercepto: {intercepto_y:.2f}', xy=(0, intercepto_y), xytext=(0.05, intercepto_y + 0.05),
              arrowprops=dict(arrowstyle='->'))
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plotar_composicoes(xD, zF, xB)
plotar_LOR(R)
plt.show()
```

## 4 Linha de AlimentaГ§ГЈo (Linha-q)

A linha-q define o estado tГ©rmico da carga. Ela parte do ponto (xFвЂ‹,xFвЂ‹) na diagonal.



A inclinaГ§ГЈo Г© dada por $$\frac{q}{1 - qвЂ‹}$$

Para tornar a Reta q dinГўmica, precisamos calcular o seu coeficiente angular a partir do valor de q. A equaГ§ГЈo da reta q Г©:
$$y=\frac{q}{qв€’1}вЂ‹x-\frac{zF}{q-1}вЂ‹$$

Essa reta sempre passa pelo ponto (zFвЂ‹,zFвЂ‹) na diagonal. O desafio na programaГ§ГЈo Г© que, quando q=1 (lГ­quido saturado), o denominador se torna zero , o que causa um erro no Python.

```python
def plotar_q_line(q):

  # 1. Defina o valor de q (Estado tГ©rmico da alimentaГ§ГЈo)
  # q = 1.0 (LГ­quido saturado - Vertical)
  # q = 0.0 (Vapor saturado - Horizontal)
  # q > 1.0 (LГ­quido sub-resfriado)
  # 0 < q < 1 (Mistura lГ­quido-vapor)
  # q < 0.0 (Vapor superaquecido)

  # 2. LГіgica para plotar a reta q
  if q == 1:
      # Se q=1, a reta Г© uma vertical perfeita do zF atГ© a curva de equilГ­brio
      # (Para fins didГЎticos, vamos traГ§ГЎ-la atГ© y=0.8 apenas para visualizaГ§ГЈo)
      plt.plot([zF, zF], [zF, 0.8], 'g-', # green, sГіlida
               linewidth=2, label='Reta q (q=1)')
  else:
      # Para qualquer outro valor de q, calculamos dois pontos:
      # Ponto 1: (zF, zF) - Onde ela nasce na diagonal
      # Ponto 2: Um ponto auxiliar (x_aux) para definir a direГ§ГЈo
      x_aux = zF - 0.3 # Pequeno deslocamento para a esquerda
      y_q = (q / (q - 1)) * x_aux - (zF / (q - 1))

      plt.plot([zF, x_aux], [zF, y_q], 'g-', linewidth=2, label=f'Reta q (q={q})')

  plt.legend()
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plotar_composicoes(xD, zF, xB)
plotar_LOR(R)
plotar_q_line(q)
plt.show()
```

## 5 Linha de OperaГ§ГЈo da SeГ§ГЈo de Esgotamento (LOE)

Esta linha representa a parte inferior da coluna.

Ela deve ligar o ponto (xBвЂ‹,xBвЂ‹) na diagonal atГ© o ponto de intersecГ§ГЈo entre a LOR e a Linha-q.  

Como calculamos o ponto de interseГ§ГЈo?  

1. A EquaГ§ГЈo da Reta

Toda reta pode ser escrita como:
$$y=ax+b$$

    a Г© a inclinaГ§ГЈo (coeficiente angular).

    b Г© onde ela corta o eixo vertical (intercepto).

No cГіdigo, definimos os coeficientes para as duas retas:

  LOR:
  $$y = \frac{R}{R+1}x + \frac{x_D}{R+1}$$
  
  $$a_1вЂ‹=\frac{R}{R+1}вЂ‹$$ e
  $$b_1вЂ‹=\frac{x_D}{R+1}вЂ‹вЂ‹$$

  q-line:  
  $$y=\frac{q}{qв€’1}вЂ‹x-\frac{zF}{q-1}вЂ‹$$
  
  $$a_2вЂ‹=\frac{q}{q-1}$$вЂ‹ e
  $$b_2вЂ‹=\frac{-z_F}{q-1}$$

2. O Ponto de Encontro (xiвЂ‹,yiвЂ‹)

No ponto onde elas se cruzam, o valor de y da primeira reta deve ser igual ao valor de y da segunda reta. EntГЈo, igualamos as duas:
$$a_1вЂ‹в‹…x_iвЂ‹+b_1вЂ‹=a_2вЂ‹в‹…x_iвЂ‹+b_2вЂ‹$$

Agora, o objetivo Г© isolar o $$x_i$$вЂ‹ (a coordenada X da interseГ§ГЈo):

Passamos os termos com x para um lado: $$a_1 вЂ‹x_iвЂ‹в€’a_2вЂ‹x_iвЂ‹=b_2вЂ‹в€’b_1вЂ‹$$

Colocamos xiвЂ‹ em evidГЄncia: $$x_iвЂ‹(a_1вЂ‹в€’a_2вЂ‹)=b_2вЂ‹в€’b_1вЂ‹$$

Isolamos o xiвЂ‹:

$$x_iвЂ‹=\frac{b_2вЂ‹в€’b_1вЂ‹вЂ‹}{a_1вЂ‹в€’a_2вЂ‹}$$

No cГіdigo, essa Г© a linha:  

    xi = (b2 - b1) / (a1 - a2).


3. E como achamos o yiвЂ‹?

Depois que vocГЄ jГЎ sabe o valor de xiвЂ‹, basta "jogar" esse valor em qualquer uma das duas equaГ§Гµes de reta para descobrir a altura (y). No cГіdigo, usamos a da LOR:  

$$y_i = a_1 * x_i + b_1$$  

O caso especial:

    if  q == 1

Quando a alimentaГ§ГЈo Г© lГ­quido saturado (q=1), a reta q Г© totalmente vertical e passa por zFвЂ‹.

  Nesse caso, nГЈo precisamos de conta complexa: o xiвЂ‹ Г© o prГіprio zFвЂ‹.

  O cГіdigo apenas calcula qual Г© o valor de y na LOR quando x=zFвЂ‹.

```python
def plotar_LOE(xB):

  # 1. Encontrar matematicamente onde a LOR e a Reta q se cruzam (Ponto de InterseГ§ГЈo)
  # Coeficientes da LOR (y = a1*x + b1)
  a1 = R / (R + 1)
  b1 = xD / (R + 1)

  # Coeficientes da Reta q (y = a2*x + b2)
  if q == 1:
      xi = zF
      yi = a1 * xi + b1
  else:
      a2 = q / (q - 1)
      b2 = -zF / (q - 1)
      # Ponto de encontro: xi = (b2 - b1) / (a1 - a2)
      xi = (b2 - b1) / (a1 - a2)
      yi = a1 * xi + b1

  # 2. Plotar a LOE
  # Ela liga o ponto (xB, xB) atГ© o ponto de interseГ§ГЈo (xi, yi)
  plt.plot([xB, xi], [xB, yi], color='blue', linewidth=1, label='LOE (Esgotamento)')

  # 3. Marcar o ponto de interseГ§ГЈo para clareza
  plt.plot(xi, yi, 'ko', markersize=5)
  plt.annotate(f"Ponto de\nInterseГ§ГЈo\n({round(xi,2)},{round(yi,2)})", xy=(xi, yi), xytext=(xi + 0.06, yi - 0.2),
              arrowprops=dict(arrowstyle='->', color='black'))

  plt.legend()
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plotar_composicoes(xD, zF, xB)
plotar_LOR(R)
plotar_q_line(q)
plotar_LOE(xB)
plt.show()
```

## 6 Limpeza no cГіdigo e no grГЎfico

```python
import matplotlib.pyplot as plt
import numpy as np

def plotar_grafico(xD, zF, xB, R, q):

  # 1. Curva de EquilГ­brio e Diagonal
  x_eq = np.linspace(0, 1, 100)
  y_eq = (alfa * x_eq) / (1 + (alfa - 1) * x_eq)

  plt.figure(figsize=(8, 8))
  plt.plot(x_eq, y_eq, 'b', label='EquilГ­brio') # curva de equilГ­brio
  plt.plot([0, 1], [0, 1], 'k--', alpha=0.5) # linha diagonal

  # 2. CГЎlculos da InterseГ§ГЈo (Onde as retas se encontram)
  # Coeficientes da LOR
  a_lor = R / (R + 1)
  b_lor = xD / (R + 1)

  if q == 1: # lembrando que quando q=1, o coefieciente angular da equaГ§ГЈo da q-line tende ao infinito
      xi = zF
      yi = a_lor * xi + b_lor
  else:
      a_q = q / (q - 1) #coeficiente angular da q-line
      b_q = -zF / (q - 1) #coeficiente linear da q-line
      xi = (b_q - b_lor) / (a_lor - a_q) # cГЎlculo de xi do ponto de encontro das equaГ§ГЈo LOR e q-line
      yi = a_lor * xi + b_lor # cГЎlculo de yi do ponto de encontro das duas equaГ§Гµes

  # 3. Plotando as 3 Linhas de OperaГ§ГЈo (Todas terminam em xi, yi)

  # LOR: do ponto (xD, xD) atГ© o encontro (xi, yi)
  plt.plot([xD, xi], [xD, yi], 'r', linewidth=1, label='LOR')
  plt.plot([xD, xD], [0, xD], color='red', linestyle=':', linewidth=1) # linha vertical

  # Reta q: do ponto (zF, zF) atГ© o encontro (xi, yi)
  if q == 1:

    plt.plot([zF, xi], [zF, yi], 'g', linewidth=1, label='Reta q (q=1)')

  else:

    plt.plot([zF, xi], [zF, yi], 'g', linewidth=1, label=f"Reta q (q={q})")

  plt.plot([zF, zF], [0, zF], color='green', linestyle=':', linewidth=1) # linha vertical

  # LOP: do ponto (xB, xB) atГ© o encontro (xi, yi)
  plt.plot([xB, xi], [xB, yi], 'm', linewidth=1, label='LOE')
  plt.plot([xB, xB], [0, xB], color='m', linestyle=':', linewidth=1) # linha vertical

  # 4. Marcando as composiГ§Гµes no eixo X
  plt.plot([xD, zF, xB], [0, 0, 0], 'ko', markersize=4)

  # ConfiguraГ§Гµes Finais
  plt.xlim(0, 1)
  plt.ylim(0, 1)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Diagrama de McCabe-Thiele - Linhas de OperaГ§ГЈo')
  plt.grid(True, alpha=0.3) # Corrected 'alfa' to 'alpha'
  plt.legend()

  # Retorna dicionГЎrio com dados necessГЎrios para a prГіxima etapa
  return {
      'xi': xi, 'yi': yi, 'a_lor': a_lor, 'b_lor': b_lor,
      'xD': xD, 'xB': xB, 'zF': zF, 'alfa': alfa
  }
```

```python
plotar_grafico(xD, zF, xB, R, q)
plt.show()
```

## 7 Plotando os degraus

```python
def plotar_degraus(dados):

    # ExtraГ§ГЈo de variГЎveis do dicionГЎrio
    xi, yi = dados['xi'], dados['yi']
    a_lor, b_lor = dados['a_lor'], dados['b_lor']
    xD, xB, zF, alfa = dados['xD'], dados['xB'], dados['zF'], dados['alfa']

    x_atual = xD
    y_atual = xD
    estagios = 0
    prato_alimentacao = 0

    # LaГ§o de ConstruГ§ГЈo dos Degraus
    while x_atual > xB:
        estagios = estagios + 1

        # Passo Horizontal (EquilГ­brio)
        x_equilibrio = y_atual / (alfa - y_atual * (alfa - 1))
        plt.plot([x_atual, x_equilibrio], [y_atual, y_atual], 'k', linewidth=1) #plota a linha horizontal atГ© a curva de equilГ­brio

        x_atual = x_equilibrio

        # Passo Vertical (Linha de OperaГ§ГЈo correspondente)
        if x_atual > xi:
            y_proximo = a_lor * x_atual + b_lor # utiliza a reta LOR para calcular o y

        else: # Em algum momento, x_atual serГЎ menor que xi, e o algoritmo entrarГЎ neste "else" para marcar o prato de alimentaГ§ГЈo

          # Localizando o Prato de AlimentaГ§ГЈo
            if prato_alimentacao == 0:
                prato_alimentacao = estagios
                plt.plot(x_atual, y_atual, 'go', markersize=8, label='Prato de AlimentaГ§ГЈo') #plota a localizaГ§ГЈo do prato de alimentaГ§ГЈo

            # CГЎlculo do coeficiente da LOE para o ponto atual
            a_loe = (yi - xB) / (xi - xB)
            b_loe = xB - a_loe * xB
            y_proximo = a_loe * x_atual + b_loe #EquaГ§ГЈo da LOE

        plt.plot([x_atual, x_atual], [y_atual, y_proximo], 'k', linewidth=1) #plota a linha vertical atГ© a LOR ou LOE

        y_atual = y_proximo

    # ExibiГ§ГЈo de InformaГ§Гµes no GrГЎfico (Caixa de Texto)
    info_text = (f'Total de EstГЎgios = {estagios}\n'
          f'Prato de AlimentaГ§ГЈo = {prato_alimentacao}\n'
          f'RazГЈo de Refluxo (R) = {R}\n'
          f'q = {q}'
                 )

    plt.text(0.6, 0.15, info_text, fontsize=12, verticalalignment='bottom',
      bbox=dict(boxstyle='round,pad=0.5', facecolor='white', alpha=0.8, edgecolor='gray'))
```

```python
# Chamada sequencial das funГ§Гµes
dados_calculados = plotar_grafico(xD, zF, xB, R, q)
plotar_degraus(dados_calculados)
plt.show()
```
