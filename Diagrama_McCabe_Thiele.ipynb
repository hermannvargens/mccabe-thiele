## Diagrama de McCabe-Thiele

Dados do Problema

    Alimentação (zF​): 0,5 (50% do componente mais volátil).

    Destilado desejado (xD​): 0,9.

    Resíduo desejado (xB​): 0,1.

    Razão de Refluxo (R): 2,0.

    Condição da Carga (q): Líquido saturado (q=1).

    Equilíbrio: volatilidade relativa α=2,5.

## 0 Definir as variáveis do problema

```python
xD, zF, xB = 0.95, 0.5, 0.05
R = 3.5
q = 1
alfa = 2.5
```

## 1 Definir a Curva de Equilíbrio

A curva de equilíbrio para sistemas ideais segue a equação:

$$y = \frac{\alpha \cdot x}{1 + (\alpha - 1)x}$$

Para $\alpha = 2,5$, se $x = 0,4$, então:

$$y = \frac{2,5 \cdot 0,4}{1 + (1,5 \cdot 0,4)} = 0,625$$

Você plota esses pontos para formar a curva acima da diagonal de 45°.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vamos criar um array com todos os valores de x para os quais queremos encontrar y, e construir a curva de equilíbrio:

x_eq = np.linspace(0, 1, 100)

# Em seguida, calculamos os valores de y correspondentes:

y_eq = (alfa * x_eq) / (1 + (alfa - 1) * x_eq)

# Vamos ver alguns valores:

print("Valores de x:\n", x_eq[:10])
print("\nValores de y:\n", y_eq[:10])
```

```python
# Agora vamos ver como ficou o gráfico da curva de equilíbrio:

# A curva de equilíbrio
plt.plot(x_eq, y_eq, label='Curva de Equilíbrio')
plt.title('Diagrame de McCabe-Thiele')
plt.legend()
plt.grid(True)

#Para visualizarmos também a reta diagonal, precisamos criá-la:

plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)

plt.show()
```

A partir daqui, vamos utilizar funções para modularizar cada etapa que estamos adicionando ao gráfico.

```python
# Função para a Base do Gráfico
def plotar_base_equilibrio(x_eq, y_eq):
    plt.plot(x_eq, y_eq, label='Curva de Equilíbrio')
    plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)
    plt.title('Diagrama de McCabe-Thiele')
    plt.legend()
    plt.grid(True)
    plt.xlim(0, 1)
    plt.ylim(0, 1)
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plt.show()
```

## 2 Localização das Composições

Marque no eixo x (fase líquida) as três frações molares principais:

    xD​: Fração desejada no destilado (topo).

    xF​: Fração na alimentação (feed).

    xB​: Fração no resíduo (fundo).

```python
def plotar_composicoes(xD, zF, xB):

    # Para visualizarmos também a reta diagonal, precisamos criá-la:

    plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)

    # Ponto do Destilado (xD) no eixo x
    plt.plot([xD, xD], [0, xD], color='red', linestyle=':', linewidth=1) # linha vertical
    plt.plot(xD, 0, 'ro', clip_on=False) # 'ro' = círculo vermelho; clip_on=False permite desenhar na borda
    plt.annotate('$x_D$', xy=(xD, 0), xytext=(xD, -0.05),
                ha='center', fontweight='bold', color='red')

    # Ponto da Alimentação (zF) no eixo x
    plt.plot([zF, zF], [0, zF], color='green', linestyle=':', linewidth=1)
    plt.plot(zF, 0, 'go', clip_on=False) # 'go' = círculo verde
    plt.annotate('$z_F$', xy=(zF, 0), xytext=(zF, -0.05),
                ha='center', fontweight='bold', color='green')

    # Ponto do Fundo (xB) no eixo x
    plt.plot([xB, xB], [0, xB], color='blue', linestyle=':', linewidth=1)
    plt.plot(xB, 0, 'bo', clip_on=False) # 'bo' = círculo azul
    plt.annotate('$x_B$', xy=(xB, 0), xytext=(xB, -0.05),
                ha='center', fontweight='bold', color='blue')

    # --- DETALHES VISUAIS ---

    plt.xlabel('Fração molar no líquido (x)')
    plt.ylabel('Fração molar no vapor (y)')
    plt.grid(True)
    plt.legend()
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plotar_composicoes(xD, zF, xB)
plt.show()
```

## 3 Linha de Operação da Seção de Retificação (LOR)

Esta linha representa o que acontece acima do prato de alimentação. Ela parte do ponto $(x_D, x_D)$ na linha de 45°.

A equação é:

$$y = \frac{R}{R+1}x + \frac{x_D}{R+1}$$

Onde $R$ é a razão de refluxo. Você pode traçá-la ligando $(x_D, x_D)$ ao intercepto no eixo $y$ dado por $\frac{x_D}{R+1}$.

```python
def plotar_LOR(R):

  # 1. Calculando o intercepto no eixo y (ponto onde x = 0)
  intercepto_y = xD / (R + 1)

  # 2. Plotando a LOR
  # Vamos ligar o ponto (0, intercepto_y) até o ponto (xD, xD)
  plt.plot([0, xD], [intercepto_y, xD], color='red', linewidth=1, label='LOR (Retificação)')
  plt.legend()

  # 3. Marcando o ponto do intercepto
  plt.plot(0, intercepto_y, 'ko') # Ponto preto no eixo y
  plt.annotate(f'Intercepto: {intercepto_y:.2f}', xy=(0, intercepto_y), xytext=(0.05, intercepto_y + 0.05),
              arrowprops=dict(arrowstyle='->'))
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plotar_composicoes(xD, zF, xB)
plotar_LOR(R)
plt.show()
```

## 4 Linha de Alimentação (Linha-q)

A linha-q define o estado térmico da carga. Ela parte do ponto (xF​,xF​) na diagonal.



A inclinação é dada por $$\frac{q}{1 - q​}$$

Para tornar a Reta q dinâmica, precisamos calcular o seu coeficiente angular a partir do valor de q. A equação da reta q é:
$$y=\frac{q}{q−1}​x-\frac{zF}{q-1}​$$

Essa reta sempre passa pelo ponto (zF​,zF​) na diagonal. O desafio na programação é que, quando q=1 (líquido saturado), o denominador se torna zero , o que causa um erro no Python.

```python
def plotar_q_line(q):

  # 1. Defina o valor de q (Estado térmico da alimentação)
  # q = 1.0 (Líquido saturado - Vertical)
  # q = 0.0 (Vapor saturado - Horizontal)
  # q > 1.0 (Líquido sub-resfriado)
  # 0 < q < 1 (Mistura líquido-vapor)
  # q < 0.0 (Vapor superaquecido)

  # 2. Lógica para plotar a reta q
  if q == 1:
      # Se q=1, a reta é uma vertical perfeita do zF até a curva de equilíbrio
      # (Para fins didáticos, vamos traçá-la até y=0.8 apenas para visualização)
      plt.plot([zF, zF], [zF, 0.8], 'g-', linewidth=2, label='Reta q (q=1)')
  else:
      # Para qualquer outro valor de q, calculamos dois pontos:
      # Ponto 1: (zF, zF) - Onde ela nasce na diagonal
      # Ponto 2: Um ponto auxiliar (x_aux) para definir a direção
      x_aux = zF - 0.3 # Pequeno deslocamento para a esquerda
      y_q = (q / (q - 1)) * x_aux - (zF / (q - 1))

      plt.plot([zF, x_aux], [zF, y_q], 'g-', linewidth=2, label=f'Reta q (q={q})')

  plt.legend()
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plotar_composicoes(xD, zF, xB)
plotar_LOR(R)
plotar_q_line(q)
plt.show()
```

## 5 Linha de Operação da Seção de Esgotamento (LOE)

Esta linha representa a parte inferior da coluna.

Ela deve ligar o ponto (xB​,xB​) na diagonal até o ponto de intersecção entre a LOR e a Linha-q.  

Como calculamos o ponto de interseção?  

1. A Equação da Reta

Toda reta pode ser escrita como:
$$y=ax+b$$

    a é a inclinação (coeficiente angular).

    b é onde ela corta o eixo vertical (intercepto).

No código, definimos os coeficientes para as duas retas:

  LOR: $$a_1​=R+\frac{R}{R+1}​$$ e
  $$b_1​=\frac{x_D}{R+1}​​$$

  q-line: $$a_2​=\frac{q}{q-1}$$​ e
  $$b2​=\frac{-z_F}{q-1}$$

2. O Ponto de Encontro (xi​,yi​)

No ponto onde elas se cruzam, o valor de y da primeira reta deve ser igual ao valor de y da segunda reta. Então, igualamos as duas:
$$a_1​⋅x_i​+b_1​=a_2​⋅x_i​+b_2​$$

Agora, o objetivo é isolar o $$x_i$$​ (a coordenada X da interseção):

Passamos os termos com x para um lado: $$a_1 ​x_i​−a_2​x_i​=b_2​−b_1​$$

Colocamos xi​ em evidência: $$x_i​(a_1​−a_2​)=b_2​−b_1​$$

Isolamos o xi​:

$$x_i​=\frac{b_2​−b_1​​}{a_1​−a_2​}$$

No código, essa é a linha:  

    xi = (b2 - b1) / (a1 - a2).


3. E como achamos o yi​?

Depois que você já sabe o valor de xi​, basta "jogar" esse valor em qualquer uma das duas equações de reta para descobrir a altura (y). No código, usamos a da LOR:  

$$y_i = a_1 * x_i + b_1$$  

O caso especial:

    if  q == 1

Quando a alimentação é líquido saturado (q=1), a reta q é totalmente vertical e passa por zF​.

  Nesse caso, não precisamos de conta complexa: o xi​ é o próprio zF​.

  O código apenas calcula qual é o valor de y na LOR quando x=zF​.

```python
def plotar_LOE(xB):

  # 1. Encontrar matematicamente onde a LOR e a Reta q se cruzam (Ponto de Interseção)
  # Coeficientes da LOR (y = a1*x + b1)
  a1 = R / (R + 1)
  b1 = xD / (R + 1)

  # Coeficientes da Reta q (y = a2*x + b2)
  if q == 1:
      xi = zF
      yi = a1 * xi + b1
  else:
      a2 = q / (q - 1)
      b2 = -zF / (q - 1)
      # Ponto de encontro: xi = (b2 - b1) / (a1 - a2)
      xi = (b2 - b1) / (a1 - a2)
      yi = a1 * xi + b1

  # 2. Plotar a LOE
  # Ela liga o ponto (xB, xB) até o ponto de interseção (xi, yi)
  plt.plot([xB, xi], [xB, yi], color='blue', linewidth=1, label='LOE (Esgotamento)')

  # 3. Marcar o ponto de interseção para clareza
  plt.plot(xi, yi, 'ko', markersize=5)
  plt.annotate(f"Ponto de\nInterseção\n({round(xi,2)},{round(yi,2)})", xy=(xi, yi), xytext=(xi + 0.06, yi - 0.2),
              arrowprops=dict(arrowstyle='->', color='black'))

  plt.legend()
```

```python
plotar_base_equilibrio(x_eq, y_eq)
plotar_composicoes(xD, zF, xB)
plotar_LOR(R)
plotar_q_line(q)
plotar_LOE(xB)
plt.show()
```

## Limpeza no código e no gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

def plotar_grafico(xD, zF, xB, R, q):

  # 1. Curva de Equilíbrio e Diagonal
  x_eq = np.linspace(0, 1, 100)
  y_eq = (alfa * x_eq) / (1 + (alfa - 1) * x_eq)

  plt.figure(figsize=(8, 8))
  plt.plot(x_eq, y_eq, 'b', label='Equilíbrio') # curva de equilíbrio
  plt.plot([0, 1], [0, 1], 'k--', alpha=0.5) # linha diagonal

  # 2. Cálculos da Interseção (Onde as retas se encontram)
  # Coeficientes da LOR
  a_lor = R / (R + 1)
  b_lor = xD / (R + 1)

  if q == 1: # lembrando que quando q=1, o coefieciente angular da equação da q-line tende ao infinito
      xi = zF
      yi = a_lor * xi + b_lor
  else:
      a_q = q / (q - 1) #coeficiente angular da q-line
      b_q = -zF / (q - 1) #coeficiente linear da q-line
      xi = (b_q - b_lor) / (a_lor - a_q) # cálculo de xi do ponto de encontro das equação LOR e q-line
      yi = a_lor * xi + b_lor # cálculo de yi do ponto de encontro das duas equações

  # 3. Plotando as 3 Linhas de Operação (Todas terminam em xi, yi)

  # LOR: do ponto (xD, xD) até o encontro (xi, yi)
  plt.plot([xD, xi], [xD, yi], 'r', linewidth=1, label='LOR')
  plt.plot([xD, xD], [0, xD], color='red', linestyle=':', linewidth=1) # linha vertical

  # Reta q: do ponto (zF, zF) até o encontro (xi, yi)
  if q == 1:

    plt.plot([zF, xi], [zF, yi], 'g', linewidth=1, label='Reta q (q=1)')

  else:

    plt.plot([zF, xi], [zF, yi], 'g', linewidth=1, label=f"Reta q (q={q})")

  plt.plot([zF, zF], [0, zF], color='green', linestyle=':', linewidth=1) # linha vertical

  # LOP: do ponto (xB, xB) até o encontro (xi, yi)
  plt.plot([xB, xi], [xB, yi], 'm', linewidth=1, label='LOE')
  plt.plot([xB, xB], [0, xB], color='m', linestyle=':', linewidth=1) # linha vertical

  # 4. Marcando as composições no eixo X
  plt.plot([xD, zF, xB], [0, 0, 0], 'ko', markersize=4)

  # Configurações Finais
  plt.xlim(0, 1)
  plt.ylim(0, 1)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Diagrama de McCabe-Thiele - Linhas de Operação')
  plt.grid(True, alpha=0.3) # Corrected 'alfa' to 'alpha'
  plt.legend()

  # Retorna dicionário com dados necessários para a próxima etapa
  return {
      'xi': xi, 'yi': yi, 'a_lor': a_lor, 'b_lor': b_lor,
      'xD': xD, 'xB': xB, 'zF': zF, 'alfa': alfa
  }
```

```python
plotar_grafico(xD, zF, xB, R, q)
plt.show()
```

## 6 Plotando os degraus

```python
def plotar_degraus(dados):
    """
    Executa os passos 5 e 6: Construção iterativa dos estágios teóricos
    e formatação final do gráfico.
    """
    # Extração de variáveis do dicionário
    xi, yi = dados['xi'], dados['yi']
    a_lor, b_lor = dados['a_lor'], dados['b_lor']
    xD, xB, zF, alfa = dados['xD'], dados['xB'], dados['zF'], dados['alfa']

    x_atual = xD
    y_atual = xD
    estagios = 0
    prato_alimentacao = 0

    # 5. Laço de Construção dos Degraus
    while x_atual > xB:
        estagios = estagios + 1

        # Passo Horizontal (Equilíbrio)
        x_equilibrio = y_atual / (alfa - y_atual * (alfa - 1))
        plt.plot([x_atual, x_equilibrio], [y_atual, y_atual], 'k', linewidth=1)

        x_atual = x_equilibrio

        # Passo Vertical (Linha de Operação correspondente)
        if x_atual > xi:
            y_proximo = a_lor * x_atual + b_lor


        else:

          # Localizando o Prato de Alimentação
            if prato_alimentacao == 0:
                prato_alimentacao = estagios
                plt.plot(x_atual, y_atual, 'go', markersize=8, label='Prato de Alimentação')

            # Cálculo do coeficiente da LOE para o ponto atual
            a_loe = (yi - xB) / (xi - xB)
            b_loe = xB - a_loe * xB
            y_proximo = a_loe * x_atual + b_loe

        plt.plot([x_atual, x_atual], [y_atual, y_proximo], 'k', linewidth=1)
        y_atual = y_proximo

    # 6. Exibição de Informações no Gráfico (Caixa de Texto)
    info_text = (f'Total de Estágios: {estagios}\n'
          f'Prato de Alimentação: {prato_alimentacao}\n'
          f'Razão de Refluxo (R): {R}\n'
          f'q: {q}'
                 )

    plt.text(0.6, 0.15, info_text, fontsize=12, verticalalignment='bottom',
      bbox=dict(boxstyle='round,pad=0.5', facecolor='white', alpha=0.8, edgecolor='gray'))
```

```python
# Chamada sequencial das funções
dados_calculados = plotar_grafico(xD, zF, xB, R, q)
plotar_degraus(dados_calculados)
plt.show()
```
