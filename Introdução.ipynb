## 3 Operadores Aritméticos e Precedência

A avaliação de expressões segue a ordem de precedência: Parênteses, Exponenciação, Multiplicação/Divisão e Adição/Subtração.

    ** : Exponenciação.

    // : Divisão inteira (descarta o decimal).

    %  : Resto da divisão (módulo).

O PEMDAS é um acrônimo mnemônico usado para lembrar a ordem de precedência das operações matemáticas. Tanto na álgebra manual quanto em linguagens de programação como Python, essa ordem é fundamental para não obter resultados errados.  

| Letra | Significado | Descrição |
| :--- | :--- | :--- |
| **P** | **Parênteses** | Tudo dentro de `()` é resolvido primeiro. |
| **E** | **Expoentes** | Potenciação (em Python, usa-se `**`). |
| **M / D** | **Multiplicação / Divisão** | Resolvidos da esquerda para a direita (possuem o mesmo nível de prioridade). |
| **A / S** | **Adição / Subtração** | Resolvidos por último, da esquerda para a direita. |

Vamos calcular a Razão de Refluxo Mínima para uma coluna de destilação binária:


$$R_{min} = \frac{1}{\alpha - 1} \cdot \left( \frac{x_D}{z_F} - \alpha \cdot \frac{1 - x_D}{1 - z_F} \right)$$




```python
# Dados do Problema
xD = 0.95    # Pureza desejada no topo
zF = 0.50    # Composição da carga (alimentação)
alpha = 2.5  # Volatilidade relativa (facilidade de separação)
```

```python
# Cálculo de Rmin
# Parte 1: O fator de volatilidade
fator_alpha = 1 / (alpha - 1)

# Parte 2: O termo das composições
# Usamos parênteses para garantir que as frações sejam calculadas antes da subtração
termo_comp = (xD / zF) - (alpha * (1 - xD) / (1 - zF))

# Resultado Final
rmin = fator_alpha * termo_comp

print(f"A Razão de Refluxo Mínima calculada é: {rmin:.4f}")
```

## 4 Estruturas de Dados: Listas vs. NumPy Arrays

A distinção entre coleções nativas e arranjos numéricos é fundamental para o desempenho computacional.

    Listas: Estruturas heterogêneas e dinâmicas. Operações como + realizam concatenação.

    Arrays (NumPy): Estruturas homogêneas que permitem vetorização (operações elemento a elemento).

```python
import numpy

lista_dados = [1.0, 2.0, 3.0]
array_dados = numpy.array([1.0, 2.0, 3.0])

# TODO: Multiplique ambos por 2 e observe a diferença de comportamento
res_lista = 2*lista_dados
res_array = 2*array_dados

print(f"Resultado Lista: {res_lista}")
print(f"Resultado Array: {res_array}")
```
